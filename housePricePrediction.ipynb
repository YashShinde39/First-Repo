```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train=np.array([1,2,3,4])
y_train=np.array([200,300,400,500])
```

```python
plt.scatter(x_train,y_train,marker='x',c='r')
plt.title('Area Vs Price')
plt.xlabel('Area in thousands of feets')
plt.ylabel('House prices in thousands of dollars')
plt.show()
```

```python
def compute_model_output(x,w,b):
    n = x.shape[0]
    f_wb = np.zeros(n)
    for i in range(n):
        f_wb[i] = w*x[i] + b
    return f_wb
```

```python
n = x_train.shape[0]
x_mean = sum(x_train)/n
y_mean = sum(y_train)/n
w = (sum((x_train[i]-x_mean)*(y_train[i]-y_mean) for i in range(n))) / (sum((x_train[i]-x_mean)**2 for i in range(n)))
b = y_mean - w*x_mean
```

```python
plt.plot(x_train,compute_model_output(x_train,w,b),c='b',label ='Predicted line')
plt.scatter(x_train,y_train,marker='x',c='r',label='Training data')
plt.title('Area Vs Price')
plt.xlabel('Area')
plt.ylabel('Price')
plt.legend()
plt.show()
```
