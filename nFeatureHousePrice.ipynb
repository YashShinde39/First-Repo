```python
import numpy as np

x_train=np.array([[850,2,1,15,15],[900,2,2,12,14],[1200,3,2,10,12],[1500,3,2,8,10],[1700,3,3,7,9],
                  [1800,4,3,6,8],[2000,4,3,5,7],[2200,4,4,5,7],[2500,4,4,4,6],[2700,5,4,4,5],[3000,5,5,3,5],
                  [3200,5,5,3,4],[3500,5,5,2,4],[4000,6,6,2,3],[4200,6,6,2,3],[4500,6,6,1,2],[4800,6,7,1,2],
                  [5000,7,7,1,2],[5500,7,7,1,1],[6000,8,8,1,1]])

y_train=np.array([3200000,3500000,4800000,6200000,7000000,7600000,8500000,9200000,
                  11000000,11800000,13500000,14200000,15500000,18000000,19000000,20500000,
                  22000000,23500000,26000000,29000000])

m=x_train.shape[0]
n=x_train.shape[1]
```

```python
x_mean=np.array(np.mean(x_train[:,i]) for i in range(n))
x_max=np.array(np.max(x_train[:,i]) for i in range(n))
x_min=np.array(np.min(x_train[:,i]) for i in range(n))
```

```python
x_mean = np.mean(x_train, axis=0)
x_max  = np.max(x_train, axis=0)
x_min  = np.min(x_train, axis=0)

# normalize (vectorized)
x_train = (x_train - x_mean) / (x_max - x_min)
```

```python
def f_wb(x,w,b):
    return np.dot(w,b)+b
```

```python
def costFunction(x,y,w,b):
    m=x.shape[1]
    cost=(sum((f_wb(x[i,:],w,b)-y[i])**2 for i in range(m)))/(2*m)

    return cost
```

```python
a=0.1
w=np.zeros(n)
b=0
for i in range(1000):
    prev=costFunction(x_train,y_train,w,b)
    dJ_dw=np.zeros(n)
    for j in range(n):
        dJ_dw[j]=(sum((f_wb(x_train[k,:],w,b)-y_train[j])*x_train[k,j] for k in range(m)))/m
    dJ_db=(sum((f_wb(x_train[k,:],w,b)-y_train[j]) for k in range(m)))/m
    
    for k in range(n):
        w[k]=w[k]-a*dJ_dw[k]
    b=b-a*dJ_db
    
    curr=costFunction(x_train,y_train,w,b)
    
    if abs(curr-prev)<1e-4:
        break
```
